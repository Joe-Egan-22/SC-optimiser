## Aim: To maximise linear functions

### Example 1

Maximise $z=2x + y$

subject to:

$2y + x \le 22 \\- 5y + 4x \le 15 \\ - y+4x \le 12 \\-y+4x=16$

$x, y \ge 0$

$x,y$ are decision variables, $z$ is the funciton to be optimised

```python
from scipy.optimize import linprog
```

NB: linprog() does not allow for maximisation. Must convert to minimisation problem:

$\Rightarrow -z = -2x - y$ must be minimised

Linprog() also does not allow $\ge$

```python
obj = [-2, -1] #coefficients of the objective function

lhs = [[2,1], [-5,4], [-1,4]] #coefficients of LHS (inequalities)

rhs = [22,15,12] #coefficients of RHS (inequalities)

lhs_eq = [[-1,4]] #coefficients of LHS (equalities)

rhs_eq = [16] #coefficients of RHS (equalities)
```

```python
#boundary of the decision variables

bnd = [(0, float('inf')), (0, float('inf'))]
```

NB: linprog() takes boundaries as 0 to inf by default

```python
optimisation = linprog(c = obj, A_ub = lhs, b_ub = rhs, A_eq = lhs_eq, b_eq = rhs_eq, bounds = bnd, method = 'simplex')

optimisation
```

May also use method= "interior-point" (default method), "revised simplex"

```python
#removing the equality constraints...

optimisation = linprog(c = obj, A_ub = lhs, b_ub = rhs, bounds = bnd, method = 'simplex')

optimisation
```

```python
# using "revised simplex"

optimisation = linprog(c = obj, A_ub = lhs, b_ub = rhs, bounds = bnd, method = 'revised simplex')

optimisation
```

```python
# using "interior point"
optimisation = linprog(c = obj, A_ub = lhs, b_ub = rhs, bounds = bnd, method = 'interior-point')

optimisation
```

## Real-World Example

There are three products ($x_1, x_2, x_3$) produced by manufacturing units:

- Profits made from each product are 10, 20 and 25 units respectively

- Max number of products produced each day is 80

- Raw material $M$ is used for these products. For $x_1$, 3 units of $M$ are needed, for $x_2$ we need 3 and for $x_3$ we need 4. There are 200 units of $M$ available each day

How many units of each product are needed for maximum profit?

Problem (put mathematicaly):

Maximise the function: $10x_1 + 20x_2 + 25x_3$

Manpower constraint: $x_1 + x_2 + x_3 \le 80$

Raw material constrain: $3x_1 + 2x_2 + 4x_3 \le 200$

Cannot have negative products (needs to be stated for clarity): $x_1, x_2, x_3 \ge 0$

```python
from scipy.optimize import linprog
```

Convert to minimisation problem by taking negative of objective function (linporg cannot do maximisation)

```python
obj = [-10, -20, -25]

lhs = [[1,1,1], [3,2,4]]

rhs = [80, 200]

bnd = [(0, float('inf')), (0, float('inf'))]

optimisation = linprog(c = obj, A_ub = lhs, b_ub = rhs, method = 'simplex')

optimisation
```

# Mixed Integer Programming in Python

This is a problem where some of the decision variables are constrained to take integer values

## Example problem

Maximise: $x + 8y$

Constraints:

- $x+6y \le 15$

- $x \le 4$

- $x\ge 0$

- $y \ge 0$

This is a linear optimisation problem where solutions are required to be integers

```python
!pip install ortools
```

```python
from ortools.linear_solver import pywraplp
```

```python
solver = pywraplp.Solver.CreateSolver('SCIP')

#NB Can use both maximisation and minimisation in this way
```

```python
# define decision variables

infinity = solver.infinity()
# x and y are non-negative variables
x = solver.IntVar(0.0, infinity, 'x')
y = solver.IntVar(0.0, infinity, 'y')
```

```python
# define constraints

solver.Add(x + 6 * y <= 15)

solver.Add(x <= 4)

# no need for matrix form
```

```python
# Maximise x+8 *y
solver.Maximize(x + 8 * y)
```

```python
status = solver.Solve()
```

```python
# print the result in convenient form

if status == pywraplp.Solver.OPTIMAL:
    print('Solution:')
    print('Objective value =', solver.Objective().Value())
    print('x =', x.solution_value())
    print('y =', y.solution_value())

else:
    print('The problem has no optimal solution')

```

```python
# Create the linear solver with GLOP backend 
#(linear programming solver which allows non-integer values)
solver = pywraplp.Solver.CreateSolver('GLOP')
```

```python
solver.Maximize(x + 8*y)
```

```python
status = solver.Solve()
```

```python
if status == pywraplp.Solver.OPTIMAL:
    print('Solution:')
    print('Objective value =', solver.Objective().Value())
    print('x =', x.solution_value())
    print('y =', y.solution_value())

else:
    print('The problem has no optimal solution')
    
```
